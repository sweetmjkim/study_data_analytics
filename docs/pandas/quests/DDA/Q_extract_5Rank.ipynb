```python
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rc('font', family='NanumBarunGothic') # 혹은 다른 설치한 Nanum 폰트 사용
```

```python
import pandas as pd
```

```python
# 리스트에 대한 내용 참고 : https://github.com/yojulab/study_data_analytics/blob/main/datasets/ShoppingMallDeliveryWithDate.md
df_LCOCT = pd.read_csv('/content/ShoppingMallDeliveryWithDate.csv')
```

```python
df_LCOCT.info()
```

```python
import seaborn as sns
```

```python
df_LCOCT.describe(include=object)
```

```python
sns.countplot(data=df_LCOCT, y='결제수단')
```

```python
df_LCOCT['결제수단'].value_counts()
```

```python
series_pay_up = df_LCOCT['결제수단'].value_counts()
series_pay_up.iloc[:3]
pay_up_list = series_pay_up.iloc[:3].index.to_list()
```

```python
series_pay_down = df_LCOCT['결제수단'].value_counts()
series_pay_down.iloc[-3:]
pay_down_list = series_pay_down.iloc[-3:].index.to_list()
```

```python
sns.countplot(data=df_LCOCT, y='결제수단', order=pay_up_list)
plt.title("상위 3위")
```

```python
sns.countplot(data=df_LCOCT, y='결제수단', order=pay_down_list)
plt.title("하위 3위")
```

```python
sns.countplot(data=df_LCOCT, y='결제수단', order=pay_up_list + pay_down_list)
plt.title("하위 3위")
```
