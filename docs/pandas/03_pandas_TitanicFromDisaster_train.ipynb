### 범주형 데이터 타입 판단
- unique : 고유한 값에 갯수
- top : 고유한 값에서 최상값(문자 : 오른차순 처음, 숫자 : 오름차순 끝)
- freq : 가장 많이 반복횟수

```python
df_titanic.describe(include=object)     # 범주형 파악
```
