```python
import pandas as pd
```

```python
df_SSList = pd.read_csv('/content/SpineSurgeryList.csv')          # 자료 가지고 오기
df_SSList.describe()
```

```python
df_SSList.columns             # 컬럼 종류 확인
```

```python
df_SSList[['환자ID','입원기간', '입원일자', '퇴원일자', '수술일자']]          # 내가 원하는 컬럼을 리스트로 만들기
```

## 입원 기간 구하기

```python
df_SSList['입원일자_datetime'] = pd.to_datetime(df_SSList['입원일자'], format='%Y%m%d')
df_SSList['퇴원일자_datetime'] = pd.to_datetime(df_SSList['퇴원일자'], format='%Y%m%d')
df_SSList['입원기간_datetime'] = df_SSList['퇴원일자_datetime'] - df_SSList['입원일자_datetime']
# df_SSList['입원기간_datetime']
df_SSList['입원기간_datetime_day'] = df_SSList['입원기간_datetime'].dt.days
# df_SSList['입원기간_datetime_day']
df_SSList_compose = df_SSList[['환자ID','입원일자_datetime','퇴원일자_datetime','입원기간_datetime_day']]
```

```python
df_SSList_compose.describe()        # 수치형 열에 대한 통계값 확인 (개수, 평균, 표준편차, 최소값, 25%, 50%, 75%, 최대값)
```

```python
# 최소 입원한 기간 : 0일(-2155일)
# 오래 입원한 기간 : 1125일
```

```python
df_SSList_compose_sort = df_SSList_compose.sort_values(by='입원기간_datetime_day', ascending=True)[-5:]     # 오래 입원한 5위
df_SSList_compose_sort
```

```python
import seaborn as sns
sns.barplot(data=df_SSList_compose_sort, x='입원기간_datetime_day', y='환자ID')
```

## 수술 일자를 (월, 주, 일) 단위로 분해하고 시각화

```python
df_SSList['수술일자_datetime'] = pd.to_datetime(df_SSList['수술일자'], format = "%Y%m%d")
# df_SSList['수술일자_datetime']
df_SSList['수술일자_datetime_month'] = df_SSList['수술일자_datetime'].dt.month
df_SSList['수술일자_datetime_week'] = df_SSList['수술일자_datetime'].dt.isocalendar().week
df_SSList['수술일자_datetime_day'] = df_SSList['수술일자_datetime'].dt.day
```

```python
sns.countplot(data=df_SSList,x='수술일자_datetime_month')
```

```python
sns.countplot(data=df_SSList,y='수술일자_datetime_week')
```

```python
sns.countplot(data=df_SSList,x='수술일자_datetime_day')
```
